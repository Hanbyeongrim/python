```python
import numpy as np
# 키세팅
key = "PRDT_CL_NM=&PRDT_CL_CD01=&PRDT_CL_CD02=&START_YMD=20190101&END_YMD=20190131&PRDT_NM=&DEP_PLACE=&SITE=&PLACE_SE_CD=LL1002&FD_LCT_CD=LCA000&IN_NM=&ATC_ID=&MDCD=&SRNO=&F_ATC_ID=&pageIndex=1&MENU_NO="
key_splited = [x.split("=") for x in key.split("&")]
key_splited[3][1] = "20181201"
key = "&".join(["=".join(k) for k in key_splited])

url = header + "?" + key

html = rq.get(url)
soup = BeautifulSoup(page.text, "html.parser")

css_ = '#paging a'
last_page = int(soup.select(css_)[-1]["onclick"].split("(")[1].split(")")[0])
last_page = 3
print(last_page)

#[x["title"] for x in soup.find_all("td",scope="row")]#실패:타이틀 있는 것도 있고 없는 것도 있는 듯
#[x["title"] for x in soup.select("tr > td")]#실패:

items =[x.text for x in soup.select("tr > td a")]
for page_num in range(2, last_page+1):
    print(page_num, "/ %d"%last_page)
    key_splited[-2][1] = str(page_num)
    key = "&".join(["=".join(k) for k in key_splited])
    url = header + "?" + key
    html = rq.get(url)
    soup = BeautifulSoup(html.text, "html.parser")
    items += [x.text for x in soup.select("tr > td a")]
#item 전처리는 각자
```
